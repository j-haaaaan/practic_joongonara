```python
def get_data(url):
    """
    URL을 받으면 json으로 데이터를 가져와서 pandas로 만들어주는 함수
    """
    df = pd.DataFrame(columns=["state", "title"])
    response = requests.get(url)
    dom = BeautifulSoup(response.content, "html.parser")
    titles = dom.select('.list_area > .board_box ')
#     print(titles)
    for title in titles:
        t1 = title.select_one(".tit").text
        t2 = t1.strip()
        df.loc[len(df)] = {
            "state" : title.select_one(".icon_txt").text,
            "title" : t2,
        }
    return df
```

```python
df = get_data(url)
df
```

### 전체 페이지에서 다 긁어와보자

```python
def get_data(url):
    df = pd.DataFrame(columns=["article_nbr"])
    response = requests.get(url)
    dom = BeautifulSoup(response.content, "html.parser", )
    titles = dom.select('.article-board .m-tcol-c')
#     print(titles)
    for title in titles:        
        df.loc[len(df)] = {            
            "article_nbr" : title.select_one("#span")
        }
    return df
```

```python
df = get_data(url)
df
```
